```python
class CocoDatasetManipulation():
    
    def __init__(self, dataset, batch_size, device):
        super(CocoDatasetManipulation, self).__init__()
        self.dataset = dataset
        self.dataset_len = len(dataset)
        self.batch_size = batch_size
        self.device = device
        self.avail_images = self.get_avail_images(dataset)

    @staticmethod
    def get_avail_images(dataset):
        return torch.ones((len(dataset)), dtype=bool)

    def reset_avail_images(self):
         self.avail_images = torch.ones((self.dataset_len), dtype=bool)

    def reset_random_images(self):
        img_number = torch.randint(0, self.dataset_len, (1,1))[0].item()
        self.avail_images[img_number] = True

```
